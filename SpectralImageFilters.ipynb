```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
# c
log_kernel = cv2.getGaussianKernel(3, 0.2)
log_mask = np.outer(log_kernel, log_kernel)
log_mask = cv2.Laplacian(log_mask, cv2.CV_64F)
#

# b
laplacian_mask = np.array([[0, 1, 0],
                          [1, -4, 1],
                          [0, 1, 0]])

# a
gaussian_Kernel = cv2.getGaussianKernel(5, 0.2)
gaussian_mask = np.outer(gaussian_Kernel, gaussian_Kernel)

# Load image
image = cv2.imread('/pic.jpg', cv2.IMREAD_GRAYSCALE)

#  filters
con_gaussian = cv2.filter2D(image, -1, gaussian_mask)
con_laplacian = cv2.filter2D(image, -1, laplacian_mask)
con_log = cv2.filter2D(image, -1, log_mask)

# Display
cv2_imshow(image)
cv2_imshow(con_gaussian)
cv2_imshow(con_laplacian)
cv2_imshow(con_log)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
